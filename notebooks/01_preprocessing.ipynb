### Binary Encoding

The target variable must be encoded as a numeric value to ensure compatibility with scikit-learn models and other analytical procedures.


```python
# import necessary libraries
import pandas as pd
from pathlib import Path

# load cleaned data to df
ROOT = Path().resolve().parents[0]
CLEANED_DATA_PATH = ROOT / "data" / "raw" / "bank-full.csv"
df = pd.read_csv(CLEANED_DATA_PATH, sep = ";")
df.head()
```

Explore the `target` variable

```python
print("-------------------------------------------")
print("Target variable 'y' overview")
print("-------------------------------------------")
print(df['y'].info())
print("\n")
print("-------------------------------------------")
print("Target variable 'y' unique value counts")
print("-------------------------------------------")
print(df['y'].value_counts())
print("\n")
print("-------------------------------------------")
print("Target variable 'y' null value counts")
print("-------------------------------------------")
print(df['y'].isna().sum())
```

The target variable is currently stored as an `object` type with no missing values and two unique categories: `yes` and `no`. For modeling and analysis purposes, it will be converted into a binary numeric variable:

- `yes` -> 1

- `no`  -> 0

```python
#Encode target
df['y'] = df['y'].map({'yes':1,'no':0})
```

```python
df.info()
```

The preprocessed dataset is saved to `data/processed/bank-full-cleaned-processed.csv` for use in downstream analysis and modeling.

```python
import os

CLEAN_PROCESSED_DATA_PATH = ROOT / "data" / "processed" / "bank-full-cleaned-processed.csv"
os.makedirs(os.path.dirname(CLEAN_PROCESSED_DATA_PATH), exist_ok=True)
df.to_csv(CLEAN_PROCESSED_DATA_PATH, index=False)
```
